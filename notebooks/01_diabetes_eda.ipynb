# Diabetes Dataset - Exploratory Data Analysis
## PIMA Indians Diabetes Dataset
**Source:** UCI ML Repository / Kaggle  
**Rows:** 768 | **Features:** 8 | **Target:** Outcome (0/1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Load dataset
# Download from: https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database
df = pd.read_csv('diabetes.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
# Dataset info
print(df.info())
print('\n' + '='*50)
print('Missing values:')
print(df.isnull().sum())
```

```python
# Class balance
print('Target distribution:')
print(df['Outcome'].value_counts())
print('\nPercentages:')
print(df['Outcome'].value_counts(normalize=True))

plt.figure(figsize=(6,4))
df['Outcome'].value_counts().plot(kind='bar', color=['#00C2A0', '#FF6B6B'])
plt.title('Class Distribution - Diabetes')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Check for zero values (biological impossibilities)
zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
print('Zero values in critical columns:')
for col in zero_cols:
    zero_count = (df[col] == 0).sum()
    print(f'{col}: {zero_count} ({zero_count/len(df)*100:.1f}%)')
```

```python
# Statistical summary
df.describe()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', center=0)
plt.title('Feature Correlation - Diabetes Dataset')
plt.tight_layout()
plt.show()
```

```python
# Distribution plots
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for idx, col in enumerate(df.columns[:-1]):
    axes[idx].hist(df[col], bins=30, color='#00C2A0', alpha=0.7, edgecolor='black')
    axes[idx].set_title(f'{col} Distribution')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

- Note zero values requiring imputation
