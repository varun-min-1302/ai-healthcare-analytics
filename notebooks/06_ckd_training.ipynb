# CKD Model Training - PyCaret Pipeline
## Chronic Kidney Disease Dataset
**Target AUC:** ≥ 0.85

```python
from pycaret.classification import *
import pandas as pd
import shap
```

```python
# Load dataset
df_c = pd.read_csv('kidney_disease.csv')
print(f'Dataset shape: {df_c.shape}')
print(f'Class distribution:\n{df_c["classification"].value_counts(normalize=True)}')
```

```python
# Setup with categorical features
exp_c = setup(
    data=df_c,
    target='classification',
    session_id=42,
    normalize=True,
    categorical_features=['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane'],
    verbose=False
)
```

```python
# Compare models
best_c = compare_models(sort='AUC', n_select=1)
print(pull().head())
```

```python
# Tune model
tuned_c = tune_model(best_c, optimize='AUC', n_iter=50, search_library='optuna')
print(pull())
```

```python
# Evaluate
evaluate_model(tuned_c)
```

```python
# Save model
final_c = finalize_model(tuned_c)
save_model(final_c, MODELS_PATH + 'ckd_model')
print('✓ CKD model saved')
```
