# Chronic Kidney Disease - Exploratory Data Analysis
## CKD Dataset
**Source:** UCI ML Repository  
**Rows:** 400 | **Features:** 24 | **Target:** classification (ckd/notckd)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Load dataset
# Download from: https://archive.ics.uci.edu/ml/datasets/chronic_kidney_disease
df = pd.read_csv('kidney_disease.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
# Dataset info
print(df.info())
print('\n' + '='*50)
print('Missing values:')
print(df.isnull().sum())
print('\nMissing percentage:')
print((df.isnull().sum() / len(df) * 100).sort_values(ascending=False))
```

```python
# Class balance
print('Target distribution:')
print(df['classification'].value_counts())
print('\nPercentages:')
print(df['classification'].value_counts(normalize=True))

plt.figure(figsize=(6,4))
df['classification'].value_counts().plot(kind='bar', color=['#00C2A0', '#FF6B6B'])
plt.title('Class Distribution - CKD')
plt.xlabel('Classification')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Statistical summary
df.describe()
```

```python
# Correlation heatmap (numeric features only)
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(14, 12))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', center=0)
plt.title('Feature Correlation - CKD Dataset')
plt.tight_layout()
plt.show()
```

## Key Observations
- CKD dataset has significant missing values (>40% in some columns)
- Mix of categorical and numerical features
- Requires careful preprocessing and imputation strategy
