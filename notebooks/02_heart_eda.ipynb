# Heart Disease Dataset - Exploratory Data Analysis
## Cleveland Heart Disease Dataset
**Source:** UCI ML Repository  
**Rows:** 303 | **Features:** 13 | **Target:** target (0/1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Load dataset
# Download from: https://archive.ics.uci.edu/ml/datasets/heart+disease
df = pd.read_csv('heart.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
# Dataset info
print(df.info())
print('\n' + '='*50)
print('Missing values:')
print(df.isnull().sum())
```

```python
# Class balance
print('Target distribution:')
print(df['target'].value_counts())
print('\nPercentages:')
print(df['target'].value_counts(normalize=True))

plt.figure(figsize=(6,4))
df['target'].value_counts().plot(kind='bar', color=['#00C2A0', '#FF6B6B'])
plt.title('Class Distribution - Heart Disease')
plt.xlabel('Target')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Statistical summary
df.describe()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', center=0)
plt.title('Feature Correlation - Heart Disease Dataset')
plt.tight_layout()
plt.show()
```

## Feature Descriptions
- age: age in years
- sex: (1 = male; 0 = female)
- cp: chest pain type (0-3)
- trestbps: resting blood pressure
- chol: serum cholesterol
- fbs: fasting blood sugar > 120 mg/dl
- restecg: resting ECG results
- thalach: maximum heart rate achieved
- exang: exercise induced angina
- oldpeak: ST depression
- slope: slope of peak exercise ST segment
- ca: number of major vessels
- thal: thalassemia
