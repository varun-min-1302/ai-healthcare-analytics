# Heart Disease Model Training - PyCaret Pipeline
## Cleveland Heart Disease Dataset
**Target AUC:** ≥ 0.85

```python
from pycaret.classification import *
import pandas as pd
import shap
```

```python
# Load dataset
df_h = pd.read_csv('heart.csv')
print(f'Dataset shape: {df_h.shape}')
print(f'Class distribution:\n{df_h["target"].value_counts(normalize=True)}')
```

```python
# Setup with categorical features
exp_h = setup(
    data=df_h,
    target='target',
    session_id=42,
    normalize=True,
    categorical_features=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'],
    verbose=False
)
```

```python
# Compare models
best_h = compare_models(sort='AUC', n_select=1)
print(pull().head())
```

```python
# Tune model
tuned_h = tune_model(best_h, optimize='AUC', n_iter=50, search_library='optuna')
print(pull())
```

```python
# Evaluate
evaluate_model(tuned_h)
```

```python
# Save model
final_h = finalize_model(tuned_h)
save_model(final_h, MODELS_PATH + 'heart_model')
print('✓ Heart model saved')
```
